For each aircraft,calculate the total revenue per year and the average revenue per ticket

```python
pd.read_sql_query("""
    -- Outer query to select aircraft details along with average revenue per ticket
    SELECT aircraft_code,
           ticket_count,
           total_revenue,
           total_revenue / ticket_count AS avg_revenue_per_ticket
    FROM (
        -- Inner query: for each aircraft, calculate total tickets and total revenue
        SELECT aircraft_code,
               COUNT(*) AS ticket_count,       -- Count total number of tickets for each aircraft
               SUM(amount) AS total_revenue    -- Sum up ticket amounts to get total revenue
        FROM ticket_flights
        JOIN flights ON ticket_flights.flight_id = flights.flight_id  -- Join flights table using flight_id
        GROUP BY aircraft_code               -- Group by aircraft to get per-aircraft stats
    )
""", connection)  # Use the SQLite connection to run the SQL
```

Calculate the average occupancy per aircraft

This code calculates the occupancy rate for each aircraft model using the data stored in a SQL database.
Occupancy rate tells us how full the aircraft is on average, i.e., how many seats are actually booked compared to how many are available.

To calculate this:

1. First, it finds how many seats were booked for each flight by counting the boarding passes.

2. Then, it looks at how many total seats are available for each aircraft model.

3. After that, it calculates the average number of booked seats per aircraft model.

4. Finally, it divides that average by the total number of seats in that aircraft model to get the occupancy rate.

```python
occupancy_rate = pd.read_sql_query("""  -- Run SQL query and store result in a DataFrame
SELECT
  a.aircraft_code,
  AVG(a.seats_count) AS booked_seats,        -- Calculate average booked seats per aircraft
  b.num_seats,                               -- Get total seats in each aircraft type
  AVG(a.seats_count)/b.num_seats AS occupancy_rate  -- Calculate occupancy rate
FROM
  (
    SELECT
      aircraft_code,
      flights.flight_id,
      COUNT(*) AS seats_count                -- Count of booked seats (boarding passes) per flight
    FROM boarding_passes
    JOIN flights
      ON boarding_passes.flight_id = flights.flight_id  -- Match boarding passes to flight info
    GROUP BY aircraft_code, flights.flight_id           -- Group by aircraft and flight ID
  ) AS a
JOIN
  (
    SELECT
      aircraft_code,
      COUNT(*) AS num_seats                 -- Total seats in each aircraft model
    FROM seats
    GROUP BY aircraft_code
  ) AS b
ON a.aircraft_code = b.aircraft_code        -- Join both tables on aircraft_code
GROUP BY a.aircraft_code                    -- Final grouping by aircraft type
""", connection)  # Connects to SQLite and runs the query
occupancy_rate
```

Calculate by how much the total annual turnover could increase by giving all aircrafts a 10% higher occupancy rate

```python
# Step 1: Increase the occupancy rate by 10%
occupancy_rate['Inc occupancy rate'] = occupancy_rate['occupancy_rate'] + occupancy_rate['occupancy_rate'] * 0.10
# Example: 0.75 + 0.075 = 0.825 → 10% increase

# Step 2: (Optional) Make sure pandas shows float numbers properly (not in scientific format)
pd.set_option("display.float_format", str)

# Step 3: Get current total revenue per aircraft model
total_revenue = pd.read_sql_query("""
    SELECT aircraft_code, SUM(amount) AS total_revenue
    FROM ticket_flights
    JOIN flights ON ticket_flights.flight_id = flights.flight_id
    GROUP BY aircraft_code
""", connection)
# This gives total revenue made from tickets, grouped by aircraft type

# Step 4: Calculate increased annual turnover based on the increased occupancy
occupancy_rate['Inc Annual Turnover'] = (
    total_revenue['total_revenue'] / occupancy_rate['occupancy_rate']   # Estimate full capacity revenue
) * occupancy_rate['Inc occupancy rate']                                 # Scale it to new occupancy

# Step 5: Display the updated table with increased turnover
occupancy_rate
```
